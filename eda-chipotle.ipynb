```python
%matplotlib inline
#url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
#chipo = pd.read_csv(url, sep = '\t')
chipotledf=pd.read_csv('/kaggle/input/chipotle/chipotle.tsv',sep="\t")
chipotledf.head()
```

## Coverting Data Type

```python
chipotledf['item_price']=chipotledf['item_price'].apply(lambda price:float(price[1:]))
```

## What was the most ordered item in the choice_description column?

```python
chipotledf.groupby('choice_description').quantity.sum()
```

## How much was the revenue for the period in the dataset?

```python
chipotledf.item_price.sum()
```

## How many orders were made in the period?

```python
chipotledf.order_id.drop_duplicates().count()
```

```python
chipotledf.order_id.nunique()
```

## What is the average amount per order?

```python
chipotledf['revenue'] = chipotledf['quantity'] * chipotledf['item_price']
```

```python
chipotledf.groupby('order_id').sum().revenue.mean()
```

```python
chipotledf.item_name.n
```

## Get top 5 items bought

```python
top5item=chipotledf.groupby('item_name').quantity.sum().sort_values(ascending=False)[0:5]
```

## Create a histogram of the top 5 items bought

```python
import pandas as pd
import matplotlib.pyplot as plt
top5item.plot(kind='bar')
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Most ordered Chipotle\'s Items')
```

## Number of items orderered per order price

```python
price_cnt=chipotledf.groupby('item_price').quantity.sum().reset_index()
```

```python
price_cnt.plot(kind='scatter',x='item_price',y='quantity')
```
